Let's create a comparision of percentages of judges to winners based on institution

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16,8))
winners.mfa_degree.value_counts(normalize=True)[:10].plot(kind='barh', ax=ax[0])
ax[0].set_title('Winners')
ax[0].set_xlabel('Count')
judges.mfa_degree.value_counts(normalize=True)[:10].plot(kind='barh', ax=ax[1])
ax[1].set_title('Judges')
ax[1].set_xlabel('Count')
plt.subplots_adjust(wspace=0.5)
```

Well this is interesting! Over 30% of winners come from University of Iowa and more than 40% of judges are or were associated with the University of Iowa. Is there enough correlation to test if there is a bias? Columbia comes in second for both winners and judges, then UC Irvine, University of Arizona, and Warren Wilson College are the remaining schools that appear in the top 10 of both winners and judges.
